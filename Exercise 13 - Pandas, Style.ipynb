Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframeâ€™s specific columns with different colors.

```python
# Importing Pandas and Numpy
import pandas as pd
import numpy as np
```

```python
# Creating a df of random values with 10 rows and 4 columns
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3', 'Column4']) # Converting the array in a df

# Introducing some NaN values
df.iloc[2, 1] = np.nan
df.iloc[5, 2] = np.nan
df.iloc[7, 0] = np.nan

df
```

```python
# Define a custom styling function to highlight columns containing at least one NaN
def highlight_nan_column(s):
    is_nan_column = s.isna().any()
    return ['background-color: red' if is_nan_column else '' for _ in s]

# Apply the styling function to the DataFrame columns
styled_df = df.style.apply(highlight_nan_column)

# Display the styled DataFrame
styled_df
```
