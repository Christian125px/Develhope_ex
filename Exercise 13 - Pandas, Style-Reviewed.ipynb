# Exercise 13 - Pandas, Style

Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframeâ€™s specific columns with different colors.

```python
# Importing Pandas and Numpy
import pandas as pd
import numpy as np
```

```python
# Creating a df of random values with 10 rows and 4 columns
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3', 'Column4']) # Converting the array in a df

# Introducing some NaN values
df.iloc[2, 1] = np.nan
df.iloc[5, 2] = np.nan
df.iloc[7, 0] = np.nan

df
```

```python
def highlight_columns(s, columns_to_highlight) :
    
    columns_to_highlight = columns_to_highlight.split(',')

    for column in columns_to_highlight :

        if column.lower() in s.name.lower():
            return ['background-color: brown'] * len(s)
        
    return [''] * len(s)

# User input to choose the columns to highlight
user_input = input('Choose columns to highlight separated by a comma: ')

# Using 'highlight_columns' function and user input
styled_df = df.style.apply(highlight_columns, columns_to_highlight=user_input)
styled_df
```
