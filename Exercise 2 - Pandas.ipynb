Write a Pandas program to filter the records from the world alcohol consumption dataset where the beer servings are greater than 200 lt and the life expectancy is above 70 years.

```python
import pandas as pd
```

```python
df = pd.read_csv('World_Alcohol_Dataset - MAIN.csv')
df.head()
```

```python
df.info()
```

```python
# There are one or more rows containing the special character '?'
df[df['beer_servings'].str.contains('\?')] # Checking the row(s) containing '?'
```

```python
# df.drop(df[df['beer_servings'] == '?'].index , inplace=True) # Dropping the unconsistent row
# df[df['beer_servings'].str.contains('\?')] # Checking if the unconsistent row has been removed
df = df[df['beer_servings'] != '?'] # Dropping the unconsistent row
df[df['beer_servings'].str.contains('\?')] # Checking if the unconsistent row has been removed
```

```python
df['beer_servings'] = df['beer_servings'].astype(float) # Converting the 'beer_servings' column to float
```

```python
df[(df['beer_servings'] > 200) & (df['life_expentancy'] > 70)] # Showing the df with filtering condition, we've not saved the changes in the df
```
