# Exercise 14 - Pandas, Style

Write a Pandas program to display the dataframe in Heatmap style.

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a df of random generated numbers with 10 rows and 4 columns 
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3', 'Column4']) # Converting the array in a df

# Showing the df
df
```

```python
# Displaying the df in heatmap style(No NaN values)
# plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')

plt.show()
```

```python
# Creating a df of random values with 10 rows and 4 columns
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3', 'Column4']) # Converting the array in a df

# Introducing some NaN values
df.iloc[2, 1] = np.nan
df.iloc[5, 2] = np.nan
df.iloc[7, 0] = np.nan

df
```

```python
# Displaying the df in heatmap style(With NaN values)
# plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')

plt.show()
```
