# Exercise 14 - Pandas, Style

Write a Pandas program to display the dataframe in Heatmap style.

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a df of random generated numbers with 10 rows and 4 columns 
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3', 'Column4']) # Converting the array in a df

# Showing the df
df
```

```python
df.style.background_gradient(cmap='PuBu') # basic
```

```python
df.style.background_gradient(cmap='PuBu', low=0, high=0, axis=0, subset=None, text_color_threshold=0.408, vmin=0.7, vmax=0.85, gmap=None) # more details
```

```python
df.style.background_gradient(cmap='PuBu', low=0, high=0, axis=0, subset=['Column3', 'Column4'], vmin=0.7, vmax=0.9, gmap=None) # subset
```

```python
df.style.background_gradient(cmap='PuBu', low=0.4, high=0.8, axis=0, subset=None, text_color_threshold=0.408, vmin=None, vmax=None, gmap=None) # low, high
```

```python
df.style.background_gradient(cmap='coolwarm', low=0, high=0, axis=0, subset=None, text_color_threshold=0.408, vmin=None, vmax=None, gmap=None) # Changing cmap
```
