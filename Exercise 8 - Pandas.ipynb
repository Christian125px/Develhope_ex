# Exercise 8 - Pandas

Write a Pandas program to extract specific records from the world alcohol consumption dataset, starting from the 2nd row, and selecting every 5th row thereafter.

```python
import pandas as pd
```

```python
df = pd.read_csv('World_Alcohol_Dataset - MAIN.csv')
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
# There are one or more rows containing the special character '?'
df[df['beer_servings'].str.contains('\?')] # Checking the row(s) containing '?'
```

```python
# df.drop(df[df['beer_servings'] == '?'].index , inplace=True) # Dropping the unconsistent row
# df[df['beer_servings'].str.contains('\?')] # Checking if the unconsistent row has been removed
df = df[df['beer_servings'] != '?'] # Dropping the unconsistent row
df[df['beer_servings'].str.contains('\?')] # Checking if the unconsistent row has been removed
```

```python
df['beer_servings'] = df['beer_servings'].astype(float) # Converting the 'beer_servings' column to float
```

```python
df[(df['beer_servings'] > 200) & (df['life_expentancy'] > 70)] # Showing the df with filtering condition, we've not saved the changes in the df
```

```python
df[df['wine_servings'].str.contains('\?')]
```

```python
df = df[df['wine_servings'] != '?']
df[df['wine_servings'].str.contains('\?')]
```

```python
df['wine_servings'] = df['wine_servings'].astype(float)
```

```python
columns_to_show = ['country','beer_servings', 'spirit_servings', 'life_expentancy']
df[(df['gdp_per_capita'] > 10000) & (df['wine_servings'] > 100)][columns_to_show]
```

```python
df['beer_servings'] = df['beer_servings'].astype(float)
```

```python
df[df['spirit_servings'].str.contains('\?')]
```

```python
df = df[df['spirit_servings'] != '?']
```

```python
df['spirit_servings'] = df['spirit_servings'].astype(float)
```

```python
df['wine_servings'] = df['wine_servings'].astype(float)
```

```python
df['average_consumption'] = (df['beer_servings'] + df['spirit_servings'] + df['wine_servings']) / 3
df
```

```python
columns_to_show = ['country', 'average_consumption', 'region', 'gdp_per_capita']
df[ (0.5 < df['average_consumption']) & (df['average_consumption'] < 2.5) ][columns_to_show]
```

```python
df[(df['beer_servings'] > 50) & (df['spirit_servings'] > 50) & (df['wine_servings'] > 50)]['country'].to_frame()
```

```python
columns_to_show = ['country', 'beer_servings', 'spirit_servings', 'region']
sorted_df = df.sort_values(by=['beer_servings', 'spirit_servings'], ascending=False)[columns_to_show]
sorted_df.head()
```

```python
average_life_expentancy = round(df['life_expentancy'].mean(), 2)
average_life_expentancy
```

```python
columns_to_show = ['country', 'life_expentancy', 'gdp_per_capita']
df[df['life_expentancy'] > average_life_expentancy][columns_to_show]
```
