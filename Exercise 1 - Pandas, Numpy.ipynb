Create an array in size of 4*3 and assing it to arr variable 

```python
random_arr = np.random.rand(4, 3) # 'np.random.rand' for generating an array of random numbers between 0 and 1
random_arr
```

```python
random_arr = np.random.randn(4, 3) # 'np.random.randn' for generating a normally distributed array of random numbers with mean 0 and standard deviation 1
random_arr
```

```python
random_arr = np.random.uniform(-10, 10, size=(4, 3)) 
random_arr
# 'np.random.uniform' allows you to generate random numbers that are evenly distributed in a specified range.
```

```python
import matplotlib.pyplot as plt
# Representing the histogram of the distribution
plt.hist(random_arr.flatten(), bins=20, edgecolor='k') # random_arr.flatten() is used to "flatten" the multidimensional array into a one-dimensional array so that it can be used with plt.hist()
plt.title("Distribution of Random Numbers")
plt.xlabel("Random Values")
plt.ylabel("Frequency")
plt.grid(True)

# Showing the graph
plt.show()
```

##### Create one dimensional normally distrubuted array with mean 3 standard deviation is 4 and has elements of 100

```python
from numpy import random
```

```python
s = np.random.normal(loc=3, scale=4, size=100) # loc -> mean, scale -> st. dev, size -> number of elements
s
```

```python
max(s)
```

```python
min(s)
```

#### Install matplotlib library for ploting purposes and then plot the s 
- search to find how to do

```python
import matplotlib.pyplot as plt

plt.hist(s, bins=10)
plt.show()
```

```python
# plt.hist(s, bins=10)
plt.hist(s, bins=10, edgecolor='k') # edgecolor='k' gives black color to edges
plt.title("s Distribution")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

#### Group companies by first letter(lower) and take the mean incidents_85_99 and make it a dataframe again 

```python
df2 = df.groupby(df['airline'].str[0].str.lower())['incidents_85_99'].mean().reset_index()
df2.rename(columns = {'airline' : 'airline_fist_letter'}, inplace=True)
```
