# Exercise 12 - Pandas, Style

Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
# Importing Pandas and Numpy
import pandas as pd
import numpy as np
```

```python
# Creating a df of random values with 10 rows and 4 columns
data = np.random.rand(10, 4) # Creating a 10x4 array of random values
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D']) # Converting the array in a df

# Converting some values to NaN
df.iloc[2, 1] = np.nan
df.iloc[5, 2] = np.nan
df.iloc[7, 0] = np.nan

# Showing the df
df
```

```python
# Defining a custom styling function to highlight NaN values
def highlight_nan(val):
    if pd.isna(val):
        return 'background-color: red' # Changing color to a cell if it contains a null value
    # else:
    #     return ''

# Applying the styling function to the df
styled_df = df.style.applymap(highlight_nan)

# Showing the styled df
styled_df
```
